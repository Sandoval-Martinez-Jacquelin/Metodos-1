```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Función cuadrática.
def f(x):
    return (x+1)**(1/2)
```

```python
def g(x):
    return 1+(x/2)
```

```python
def h(x):
    return 1+(x/2)+(-x**2/8)
```

```python
def j(x):
    return 1+(x/2)+(-x**2/8)+(x**3/16)
```

```python
# Valores del eje X que toma el gráfico.
x=np.linspace(-10, 10,100)
```

```python
# Graficar ambas funciones.
plt.figure(figsize=(10,10))

plt.plot(x, f(x),label="f(x)=(x+1)^(1/2)")
plt.plot(x,g(x),label="P1(x)=1+(x/2)")
plt.plot(x,h(x),label="P2(x)=1+(x/2)+(-x^2/8)")
plt.plot(x,j(x),label="P3(x)=1+(x/2)+(-x^2/8)+(x^3/16)")
plt.axhline(0, color="black")
plt.axvline(0, color="black")
plt.xlim(-10, 10)
plt.ylim(-10, 10)


plt.xlabel("Eje x")
plt.ylabel("Eje y")
plt.title("Polinomio de Taylor")
plt.legend(loc = "upper left")
plt.show()
```
